```python
# create a tensor
s1= torch.tensor([7,8])

print(s1)
print(s1+7) # `+` or `torch.add` adds the number to all the elements in the tensor
print(s1-7) # `-` or `torch.sub` substracts the number from all the elements in the tensor
print(s1*2) # `*` or `tensor.mul` multiplies the number to all the elements in the tensor 
print(s1/2) # `/` or `tensor.div` divides all the elements in the tensor by the  given number

```

# Matrix multiplication:
### It is of two types :
- tensor by numbers which we discussed above,
- element-wise, 
- another is matrix-multiplication (dot product) 

```python
# element-wise multiplication
e1=torch.tensor([5,5,6,9])
e2=torch.tensor([2,2,8,8])

#now if we write e1*e2, we get each element multiplied to its corresponding element from the other tensor
print(e1*e2)
```

```python
# matrix-multiplication (using same e1 and e2 tensors from above)

print(e1 @ e2) # `@` or `tensor.matmul()` or `torch.mm()` how come? -> 5*5+5*5+6*8+9*8
```

```python
# transpose function (using same e1 and e2 tensors from above)
t1=torch.rand(3,2)
t2=torch.rand(3,2)

tt=t1.T # `tensor.T` to transpose a tensor
print(t1,tt)
```

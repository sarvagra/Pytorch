```python
# check for gpu access with pytorch:
torch.cuda.is_available()
```

```python
# set up device agnostic code : use gpu if available, else default to cpu
device="cuda" if torch.cuda.is_available() else "cpu"
device
```

```python
#count the number of devices- when using multiple gpus 
torch.cuda.device_count()
```

```python
# putting tensors on gpu:
tensor=torch.tensor([1,2,3], device='cpu')
print(tensor,"is on",tensor.device )

#move tensor to gpu if availabe:
gtensor=tensor.to(device)
print(gtensor,"is on",gtensor.device )
```

```python
# certain libraries work only on cpu such as numPy
gtensor.numpy() #see this throws an error asking to put the tensor into cpu before using numpy with it 
```

```python
# so putting a tensor back to cpu and using numpy with it :
ctensor=gtensor.cpu().numpy()
ctensor #converted to numpy array 
```

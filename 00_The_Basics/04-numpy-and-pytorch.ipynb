# Numpy and Pytorch Tensors
- numpy is the popular mathemtical computation library for python, thus pytorch has compacity to interact with it.
- data in numpy to ptorch tensor.
- numpy -> tensor -> torch.from_numpy()
- pytorch tensor -> numpy -> tensor.numpy()


```python
# importing
try:
    import torch
    import numpy
    import matplotlib.pyplot as plt
    print("Done Successfully")
except Exception as e:
    print("Failed to import",e)
```

```python
# numpy -> tensor
arr=numpy.arange(1.0,11.0) # create an array in numpy :
tensor=torch.from_numpy(arr)
arr,tensor #notice the default dtype of the array is float64 unless specified

tensor_f32=tensor.type(torch.float32)
print(tensor_f32)
```

```python
# experimenting
arr+=1
print(arr,"\n",tensor) #notice how both change; meaning a shared memory

# if we do arr= arr-1
arr= arr-1
print(arr,"\n",tensor) # notice how the array changes but tensor remains same .. thats the major difference in a+=n and a=a+n
```

```python
# tensor -> numpy
tensor= torch.zeros(7)
print(tensor) # notice default dtype float32

# convert this into numpy array :
arr=tensor.numpy()
arr # same data type as tensor 
```

```python
# experimeting 
tensor+=1 # updation follows the shared memory rule 
print(tensor,arr)

tensor=tensor-1 # redeclaration breaks the shared meemory rule 
print(tensor,arr)
```

# Tensor Aggregreation
### Finding:
- Min
- Max
- Mean
- Sum
- Positional Min and Max

```python
# create a tensor:
t1=torch.arange(0,100,10)
print(t1)
```

```python
# find the min 
mn=t1.min() # or `torch.min(t1)`

# find the max
mx=t1.max() # or `torch.max(t1)`

# find the sum 
sm=t1.sum() # or`torch.sum(t1)`

# find mean (imp cuz need to convert tensor into any fractional dtype)
mn=t1.float().mean()

print(mn,mx,sm,mean)
```

```python
# find the positional min & max
pmin=t1.argmin() # or `torch.argmin(t1)`
pmax=t1.argmax() # or `torch.argmax(t1)`

print(pmin, pmax)
```

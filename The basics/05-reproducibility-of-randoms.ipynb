# Reproducibility (taking out random out of ramdom)
### In short how neural network learns is :
- start with random numbers -> tensor operations -> update random numbers to make the mbetter representations of data -> again -> again -> again ...
- to reduce the randomness in pytorch and neural networks, use the concept of *random seed*
- *random seed* flavours the radomness
- 

```python
# importing
try:
    import torch
    import numpy
    import matplotlib.pyplot as plt
    print("Done Successfully")
except Exception as e:
    print("Failed to import",e)
```

```python
# create two random tensors :
t1=torch.rand([3,3]) # uses random seed which we can checy by torch.initial_seed()
t2=torch.rand([3,3])
print(t1,"\n",t2)
```

```python
# check if any of the values are equal in these two tensors 
print(t1==t2) #none 
```

```python
# lets create a random seed and use it to do the same 
RANDOM_SEED=42
torch.manual_seed(RANDOM_SEED) # set a manual seed 

t3=torch.rand([3,3])

torch.manual_seed(RANDOM_SEED) # set a manual seed again because its lost after each call of torch.rand()
t4=torch.rand([3,3])

print(t3,"\n",t4)
```

```python
# check again if any of the values are equal in these two tensors 
print(t3==t4) # notice the reproducibility
```
